**ATM SYSTEM**

```python
try:
    # Fixed balance
    balance = 5000

    # withdrawal amount
    amount = 1000

    print("Withdrawal amount:", amount)

    # Checking conditions
    if amount <= 0:
        print("Invalid amount Enter positive value.")

    elif amount > balance:
        print("Insufficient balance!")

    else:
        balance -= amount
        print("Withdrawal successful!")
        print("Withdrawn amount:", amount)

# Error handling
except ValueError:
    print("Invalid input! Only numbers allowed.")

# Runs if no error
else:
    print("Transaction completed successfully.")

# Always runs
finally:
    print("Remaining balance:", balance)
    print("Thank you for using ATM.")
```

3.Generator

Create a generator that yields number from 1 to 10 divisible by 3

```python
def divisible_by_3():
    # loop from 1 to 10
    for i in range(1, 11):
        # check here number is divisible by 3
        if i % 3 == 0:
            yield i   # return the value one by one

# calling the generator function
for num in divisible_by_3():
    print(num)   # print each value
```

 6.
 from a list

  nums = [3,55,32,10,"a", 50]

  Do -
  
   skip invalid values using exception
   create dictionary using comprehension
   key = index
   4. value = number/10

```python
nums = [3, 55, 32, 10, "a", 50]  # list is given

num_dict = {}  # Starting with an empty dictionary

for idx, num in enumerate(nums):  #Loop through list with index
    try:
        num_dict[idx] = int(num) / 10  #Convert to int and divide by 10, store in dict
    except:
        continue  #  Skip invalid values if conversion fails

print(num_dict)  # i am  Printing the final dictionary
```

Shopping cart

```python
class Cart:   # created class here
    def __init__(self):
        self.items = []  # here item will store

    ### for adding item
    def add_item(self, product, price):
        self.items.append((product, price))  # added product and price in the form of tuple
        print(product + " added with price " + str(price))

    def show_cart(self):
        print("items in cart")
        for product, price in self.items:  # here one by one item will go in list
            print(product + " - " + str(price))  # print  product and price

    def total_price(self):
        total = 0   # in starting total set as zero
        for _, price in self.items:   # from list taken price
            total += price
        print("total price " + str(total))  # total price will show


my_cart = Cart()  # creating object of class

# Adding items
my_cart.add_item("Apple", 50)
my_cart.add_item("Banana", 20)
my_cart.add_item("Milk", 35)

# Show cart and total
my_cart.show_cart()
my_cart.total_price()
```

salary system:

```python
# Employee class definition
class Employee:
    def __init__(self, name, salary):
        # Constructor to initialize employee name and salary
        self.name = name          # Store employee name
        self.salary = salary      # Store employee salary

    def calculate_bonus(self):
        # Calculate bonus based on salary
        # If salary > 50000, bonus = 10%
        # Else, bonus = 5%
        if self.salary > 50000:
            bonus = self.salary * 0.10
        else:
            bonus = self.salary * 0.05
        return bonus

    def total_salary(self):
        # Total salary = salary + bonus
        total = self.salary + self.calculate_bonus()
        return total

    def show_employee(self):
        # Display all employee details
        print("Employee Name:", self.name)
        print("Salary:", self.salary)
        print("Bonus:", self.calculate_bonus())
        print("Total Salary:", self.total_salary())


# ----- Example Usage -----

# Employee 1
emp1 = Employee("Shlok", 60000)
emp1.show_employee()

print("\n")  # Blank line for separation

# Employee 2
emp2 = Employee("Riya", 40000)
emp2.show_employee()
```
