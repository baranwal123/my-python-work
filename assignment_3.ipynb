# Calculator

# function
# two numbers
# (+,-,*,/)

```python
def calculator(a, b):
  operation = input("Enter the operator(+,-,*,/): ")
  if operation == '+':
    return a + b
  elif operation == '-':
    return a - b
  elif operation == '*':
    return a * b
  elif operation == '/':
    return a / b
  else:
    return 'Invalid operation'

calculator(1, 2)
```

Student Grade function

Take marks as input and return grade

marks >= 90 A
 >= 75 b
 >= 50 c
 else fail

```python
def find_grade(score):
  if score >= 90:
      grade = "A"
  elif score >= 75:
      grade = "B"
  elif score >= 50:
      grade = "C"
  else:
      grade = "Fail"

  return grade

score = int(input("Enter your score: "))
print(f"Your grade is: {find_grade(score)}")
```

Discount Cart Application

 cart =

add items to cart (name, price, Qty)
 View Cart
 apply discount
      # total >= 1000 => 20%
      # total >= 500 => 10%
      # 0
 cal Total bill
 Exit

```python
cart = []

while True:
    print("\n--- Discount Cart Menu ---")
    print("1. Add Item")
    print("2. View Cart")
    print("3. Calculate Bill")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter item name: ")
        price = float(input("Enter price: "))
        qty = int(input("Enter quantity: "))
        cart.append((name, price, qty))
        print("Item added!")

    elif choice == "2":
        print("\n--- Your Cart ---")
        if len(cart) == 0:
            print("Cart is empty")
        else:
            for item in cart:
                print("Name:", item[0], "| Price:", item[1], "| Qty:", item[2])

    elif choice == "3":
        total = 0

        for item in cart:
            total += item[1] * item[2]

        if total >= 1000:
            discount = total * 0.20
        elif total >= 500:
            discount = total * 0.10
        else:
            discount = 0

        final = total - discount

        print("Total Bill:", total)
        print("Discount:", discount)
        print("Final Bill:", final)

    elif choice == "4":
        print("Thank you! Visit again.")
        break

    else:
        print("Invalid choice")
```

```python
def is_prime(n):
  for i in range(2, n):
    if n%i == 0:
      return False
    else:
      return True
n = int(input())
if is_prime(n):
  print('Given Number is Prime')
else:
  print('Given Number is not Prime')
```
