4.  Given a list of numbers, count how many are even and how many are odd

```python
numbers = [1, 2, 3, 4, 5, 6]

even = 0
odd = 0

for num in numbers:
    if num % 2 == 0:
        even = even + 1
    else:
        odd = odd + 1

print("Even:", even)
print("Odd:", odd)
```

8.   Write a program to count vowels and consonants in a string.

20.   Scenario: Password strength checker – check if password contains uppercase, lowercase, digit and special character.

```python
password = input("Enter your password: ")

upper = lower = digit = special = False

for c in password:
    if c.isupper():
        upper = True
    elif c.islower():
        lower = True
    elif c.isdigit():
        digit = True
    else:
        special = True

if upper and lower and digit and special:
    print("Strong password")
else:
    print("Weak password")
```

1.   ***ATM*** Machine System (Loops + If-Else)**bold text**


```python
# Simple ATM System

pin = 1234
balance = 5000
attempts = 3

while attempts > 0:
    entered = int(input("Enter PIN: "))
    if entered == pin:
        while True:
            print("\n1: Balance  2: Withdraw  3: Exit")
            choice = input("Choice: ")

            if choice == "1":
                print("Balance:", balance)
            elif choice == "2":
                amt = int(input("Amount: "))
                if amt > balance:
                    print("Insufficient!")
                else:
                    balance -= amt
                    print("Withdrawn. Remaining:", balance)
            elif choice == "3":
                print("Bye!")
                break
            else:
                print("Invalid!")

            if input("Another? (yes/no): ").lower() != "yes":
                print("Bye!")
                break
        break
    else:
        attempts -= 1
        print("Wrong PIN. Attempts left:", attempts)

if attempts == 0:
    print("ATM blocked.")
```

2....   2. Student Result Analysis System (Loops + If-Else)
A teacher wants to analyze student marks. Requirements:

• Take marks of 5 students (input from user).
• Calculate:
o Total marks
o Average marks
o Highest marks
o Number of students passed (>= 35)
• Display grade for each student:
o =90 → A, >=75 → B, >=50 → C, else Fail

```python

marks = []
grades = []
# Loop to take marks input from user
for i in range(5):
    m = int(input(f"Enter marks for student {i+1}: "))
    marks.append(m)


total_marks = sum(marks)
average_marks = total_marks / len(marks)
highest_marks = max(marks)
passed_students = 0

# Calculate grade for each student and count passed students
for m in marks:
    if m >= 90:
        grade = 'A'
    elif m >= 75:
        grade = 'B'
    elif m >= 50:
        grade = 'C'
    elif m >= 35:
        grade = 'Pass'
    else:
        grade = 'Fail'

    grades.append(grade)

    if m >= 35:
        passed_students += 1


print("\n--- Student Result Analysis ---")
print("Marks:", marks)
print("Grades:", grades)
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
print("Highest Marks:", highest_marks)
print("Number of Students Passed:", passed_students)
```
