1. Create a program that asks the user for 5 numbers and prints the maximum and minimum using loops.

4.  Given a list of numbers, count how many are even and how many are odd

```python
numbers = [1, 2, 3, 4, 5, 6]

even = 0
odd = 0

for num in numbers:
    if num % 2 == 0:
        even = even + 1
    else:
        odd = odd + 1

print("Even:", even)
print("Odd:", odd)
```

5.  Write a program to reverse a string without using built-in reverse functions.

```python
# Take input from the user
text = input("Enter a string: ")

# Create an empty string to store the reversed result
reverse = ""

# Loop through each character in the string
for char in text:
    # Add the current character in front of the previous result
    reverse = char + reverse

# Print the final reversed string
print("Reversed string:", reverse)
```

8.   Write a program to count vowels and consonants in a string.

```python
text = input("Enter a string: ")

vowels = 0
consonants = 0

for c in text:
    if c.lower() in "aeiou":
        vowels += 1
    elif c.isalpha():
        consonants += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
```

13.  Create a program to find factorial of a number using while loop.

```python
num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact = fact * i
    i = i + 1

print("Factorial of", num, "is", fact)
```

20.   Scenario: Password strength checker – check if password contains uppercase, lowercase, digit and special character.

```python
password = input("Enter your password: ")

upper = lower = digit = special = False

for c in password:
    if c.isupper():
        upper = True
    elif c.islower():
        lower = True
    elif c.isdigit():
        digit = True
    else:
        special = True

if upper and lower and digit and special:
    print("Strong password")
else:
    print("Weak password")
```

1.   ***ATM*** Machine System (Loops + If-Else)**bold text**


```python
# Simple ATM System

pin = 1234
balance = 5000
attempts = 3

while attempts > 0:
    entered = int(input("Enter PIN: "))
    if entered == pin:
        while True:
            print("\n1: Balance  2: Withdraw  3: Exit")
            choice = input("Choice: ")

            if choice == "1":
                print("Balance:", balance)
            elif choice == "2":
                amt = int(input("Amount: "))
                if amt > balance:
                    print("Insufficient!")
                else:
                    balance -= amt
                    print("Withdrawn. Remaining:", balance)
            elif choice == "3":
                print("Bye!")
                break
            else:
                print("Invalid!")

            if input("Another? (yes/no): ").lower() != "yes":
                print("Bye!")
                break
        break
    else:
        attempts -= 1
        print("Wrong PIN. Attempts left:", attempts)

if attempts == 0:
    print("ATM blocked.")
```

```python
2....   2. Student Result Analysis System (Loops + If-Else)
A teacher wants to analyze student marks. Requirements:

• Take marks of 5 students (input from user).
• Calculate:
o Total marks
o Average marks
o Highest marks
o Number of students passed (>= 35)
• Display grade for each student:
o =90 → A, >=75 → B, >=50 → C, else Fail
```

```python
# Initialize lists to store student marks and grades
marks = []  # To store marks of 5 students
grades = []  # To store grade of each student

# Loop to take marks input from user
for i in range(5):
    m = int(input(f"Enter marks for student {i+1}: "))
    marks.append(m)

# Initialize variables
total_marks = sum(marks)
average_marks = total_marks / len(marks)
highest_marks = max(marks)
passed_students = 0

# Calculate grade for each student and count passed students
for m in marks:
    if m >= 90:
        grade = 'A'
    elif m >= 75:
        grade = 'B'
    elif m >= 50:
        grade = 'C'
    elif m >= 35:
        grade = 'Pass'
    else:
        grade = 'Fail'

    grades.append(grade)

    if m >= 35:
        passed_students += 1

# Display results
print("\n--- Student Result Analysis ---")
print("Marks:", marks)
print("Grades:", grades)
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
print("Highest Marks:", highest_marks)
print("Number of Students Passed:", passed_students)
```
