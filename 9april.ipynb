```python
import re   # import regex module

text = "i am shlok"   # this is our input string
pattern = r"[abc]"    # match any one character: a, b, or c

match = re.search(pattern, text)   # search for first match in the string

# if a match is found
if match:
    print("Match found")   # tells that match exists

    print("Value:", match.group())   # the matched character
    print("Start:", match.start())   # starting position of match
    print("End:", match.end())       # ending position of match

# if no match is found
else:
    print("No match found")
```

# 5. Employee ID Pattern (IT Department)

Employee IDs follow a pattern.

**Task:**
Match IDs like:

EMP00123  
EMP98765

Rules:

Starts with EMP  
Followed by exactly 5 digits

```python
import re

# Employee ID pattern
emp_pattern = r'^EMP[0-9]{5}$'

# Test IDs
employee_ids = ["EMP00123", "EMP98765", "EMP1234", "EMP123456", "EMPA1234"]

for emp in employee_ids:
    if re.match(emp_pattern, emp):
        print(emp, "✅ Valid")
    else:
        print(emp, "❌ Invalid")
```

1. Email Validation (HR System)
A company wants to validate employee emails.

Task: Write a regex to match valid emails like:

john.doe@company.com
hr_team@my-org.in

❌ Invalid:

john@.com
@company.com

```python
import re

emails = ["john.doe@company.com", "hr_team@my-org.in", "john@.com", "@company.com"]


email_pattern = r'^[\w.]+@[\w-]+\.[a-zA-Z]{2,4}$'    #Start se end tak pura email check karta hai:\
# username (letters/numbers/dot/underscore) + @ + domain (letters/numbers/dash) + dot + extension (2-4 letters) $ ensure fullemail is matching

for email in emails:
    if re.match(email_pattern, email):
        print(email, " Valid")
    else:
        print(email, " Invalid")
```

```python
import re

# Regex pattern for invoice
invoice_pattern = r'^INV-[0-9]{4}-[0-9]{4}$'

# Test invoice numbers
invoices = ["INV-2025-0001", "INV-2024-1234", "INV-202A-0001", "INV-2023-12"]

for invoice in invoices:
    if re.match(invoice_pattern, invoice):
        print(invoice, " Valid")
    else:
        print(invoice, " Invalid")
```

Indian Phone Number Validation (Customer Data)
A CRM system stores Indian phone numbers.

Task: Match valid numbers like:

9876543210
+91-9876543210
91 9876543210

```python
import re


phone_pattern = r'^(\+?91[\s]?)?[6-9][0-9]{9}$'   #[\s-] dash 6-9 india num
# Test numbers
numbers = [
    "9876543210",
    "+91-9876543210",
    "91 9876543210",
    "5876543210",
    "+9198765432100"
]

for number in numbers:
    if re.match(phone_pattern, number):
        print(number, "Valid")
    else:
        print(number, "Invalid")
```
